##IMPORTING LIBRARIES

```python
import nltk
import pandas as pd

from nltk.tokenize import word_tokenize
from transformers import BertTokenizer, GPT2Tokenizer
```

##USING THE SAMPLE DATA

The sample data used was AI generated

```python
text = "Natural Language Processing is transforming Artificial Intelligence systems."
```

##WORD TOKENIZATION

```python
nltk.download('punkt')
nltk.download('punkt_tab')

word_tokens = word_tokenize(text)
word_tokens
```

##CHARACTER TOKENIZATION

```python
char_tokens = list(text)
char_tokens[:30]  #
```

##WORD TOKENIZATION

```python

bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")

wordpiece_tokens = bert_tokenizer.tokenize(text.lower())
wordpiece_tokens

```

##BYTE PAIR ENCODING GPT 2

```python
gpt_tokenizer = GPT2Tokenizer.from_pretrained("gpt2")

bpe_tokens = gpt_tokenizer.tokenize(text)
bpe_tokens
```

##SENTENCE PIECE (SUBWORD TOKENIZATION)

```python
import sentencepiece as spm

print("SentencePiece is a subword tokenization method used in multilingual NLP models like T5 and ALBERT.")
```

##COMPARISM

```python
comparison = pd.DataFrame({
    "Method": ["Word-Level", "Character-Level", "WordPiece", "BPE"],
    "Token Count": [
        len(word_tokens),
        len(char_tokens),
        len(wordpiece_tokens),
        len(bpe_tokens)
    ]
})

comparison
```

##OBSERVATIONS

1.Word level tokenization split text into full words but can't handele unknown words.

2.Character level tokenization split text into different characters although detailed, it has a very long sequence.

3.Word piece (Bert) split words into sub words, it an efficient and widely used transformer.

4.SentencePiece is a language-independent subword model and used in multilingual models like T5.

5.BPE (GPT-2) merges frequent character pairs, has a good balance between vocabulary size and meaning


## Applied Learning Assignment 2: Tokenization Techniques Comparison

### 1. Objective

The objective of this assignment is to evaluate and compare different tokenization techniques used in Natural Language Processing (NLP). The techniques include word-level tokenization, character-level tokenization, Byte Pair Encoding (BPE), WordPiece, and SentencePiece.

---

### 2. Libraries Used

* NLTK for word-level tokenization

* Pandas for data analysis and comparison

These libraries are widely used in modern NLP systems and transformer-based architectures.

---

### 3. Steps of Implementation

1. Installed required NLP libraries.
2. Defined a sample text sentence.
3. Applied word-level tokenization using NLTK.
4. Applied character-level tokenization using Python string operations.

---

### 4. Observed Results

Word-level tokenization produced full words as tokens, making it easy to interpret but unable to handle unknown words effectively. Character-level tokenization produced a large number of tokens and provided fine-grained representation but increased sequence length significantly.

WordPiece and BPE tokenization methods broke rare words into subword units, making them more efficient and robust to unknown words. SentencePiece provided a flexible and language-independent subword representation suitable for multilingual NLP models.

---

### 5. Conclusion

Subword tokenization techniques such as WordPiece, BPE, and SentencePiece provide a balance between vocabulary size and semantic understanding. These methods are widely used in modern transformer-based models and outperform traditional word-level tokenization in handling unknown and rare words.
